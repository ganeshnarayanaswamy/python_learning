### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv). 

### Step 3. Assign it to a variable called drinks.

```python
drinks=pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv",sep = ',')
drinks.head()
```

### Step 4. Which continent drinks more beer on average?

```python
#drinks.continent.nunique()
beer_servings_continent=drinks[['wine_servings','beer_servings']].groupby(['continent']).mean()
#beer_servings_continent.loc[beer_servings_continent['beer_servings'].idxmax()]
beer_servings_continent[beer_servings_continent['beer_servings']==beer_servings_continent['beer_servings'].max()]
```

### Step 5. For each continent print the statistics for wine consumption.

```python
drinks[['wine_servings','continent']].describe()
```

### Step 6. Print the mean alcoohol consumption per continent for every column

```python
drinks_mean=drinks[['total_litres_of_pure_alcohol','continent']].groupby(['continent']).mean()
drinks_mean.rename(columns = {'total_litres_of_pure_alcohol':'avg_litres_of_pure_alcohol_per_continent'}, inplace = True) 
drinks_mean
#drinks_mean.columns=['continent','avg_litres_of_pure_alcohol_per_continent']
#drinks_mean

drinks.merge(drinks_mean, left_on='continent', right_on='continent')
```

### Step 7. Print the median alcoohol consumption per continent for every column

```python
drinks_median=drinks[['total_litres_of_pure_alcohol','continent']].groupby(['continent']).median()
drinks_median.rename(columns = {'total_litres_of_pure_alcohol':'median_litres_of_pure_alcohol_per_continent'}, inplace = True) 
#drinks_median
drinks.merge(drinks_median, left_on='continent', right_on='continent')
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
drinks['mean_spirit_servings'] = drinks.spirit_servings.mean()
drinks['min_spirit_servings'] = drinks.spirit_servings.min()
drinks['max_spirit_servings'] = drinks.spirit_servings.max()
drinks[['spirit_servings','mean_spirit_servings','min_spirit_servings','max_spirit_servings']]
```
